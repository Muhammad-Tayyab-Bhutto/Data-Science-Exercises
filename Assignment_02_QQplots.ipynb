```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statsmodels.graphics.gofplots import qqplot
```

```python
data = pd.read_excel('/content/drive/MyDrive/DataScience /DataSets/Juices.xlt')
```

```python
data.head(20)
```

```python
# ================= Frequency Table =====================
for column in data.columns:
    # Creating a Frequency Table = (Counting the number of occurances of each unique value in column and sorting them)
    frequency_table = data[column].value_counts().sort_index()
    print(f"\nFrequency Table for {column}:\n")
    print(frequency_table)

    # ================Histogram Plot=====================
    plt.figure(figsize=(10, 6))
    # (kde=True) parameter adds a kernel density estimate to plot for a smoother representation
    sns.histplot(data[column], kde=True)
    plt.title(f'Histogram for {column}')
    plt.xlabel('PH Level')
    plt.ylabel('Frequency')
    plt.show()

    # ================QQ ( Quantile-Quantile ) Plot================================
    #  QQ plot is used to assess if a dataset follows a particular theoretical distribution (in our dataset case we have normal distribution).
    qqplot(data[column], line='s')
    plt.title(f'QQ Plot for {column}')
    plt.show()

    # Fit the data to different distributions and printing goodness-of-fit test results
    distributions = ['norm', 'expon', 't', 'poisson']
    for distribution in distributions:
        if distribution == 'norm':
            params = stats.norm.fit(data[column])
        elif distribution == 'expon':
            params = stats.expon.fit(data[column])
        elif distribution == 't':
            params = stats.t.fit(data[column])
        elif distribution == 'poisson':
            params = (data[column].mean(),)

        stat, p_value = stats.kstest(data[column], distribution, args=params)
        print(f'K-S test for {column} and {distribution}: p-value = {p_value}')

```
